# Grouper

mais explicaçoes em: https://pbpython.com/pandas-grouper-agg.html

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
