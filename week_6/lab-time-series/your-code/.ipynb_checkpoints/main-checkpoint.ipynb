The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

* Augumented Dickey Fuller is a hypothesis test. Let's start by writing it:
> * H0: non-stationary
> * H1: stationery

```python
# import adfuller function
from statsmodels.tsa.stattools import adfuller
```

```python
def adf(col):
    adf = adfuller(col, maxlag=1)
    critical_adf = pd.DataFrame(adf[4], index=['t_statistic']).T
    stats = adf[0]
    return(stats, critical_adf)
```

```python
stats_user, df_adf_user = adf(sensor['userAcceleration.x'])
```

What are your conclusions from this test?

```python
# Conclusion: Since -24,6 < -2,86, 
#userAcceleration is stationary with more than 95% certainty'
print(f'adf test for userAcceleration: {round(stats_user, 3)}')
df_adf_user
```

```python
stats_rot, df_adf_rot = adf(sensor['rotationRate.x'])
```

```python
# Conclusion: Since -15.96 < -2,86, 
#userAcceleration is stationary with more than 95% certainty'
print(f'adf test for userAcceleration: {round(stats_rot, 3)}')
df_adf_rot
```
