## Treinando um modelo de regressão linear

Vamos agora começar a treinar o modelo de regressão. Precisamos primeiro dividir nossos dados em uma matriz X que contém os recursos para treinar, e uma matriz y com a variável alvo, neste caso, a coluna Preço. Vamos descartar a coluna "Adress" porque só tem informações de texto que o modelo de regressão linear não pode usar.

### Arrays X e y

```python
# select feature columns
X_cols = ['AvgAreaIncome', 'AvgAreaHouseAge', 'AvgAreaNumberRooms', 'AvgAreaNumberBedrooms', 'AreaPopulation']
```

```python
X = us_housing[X_cols]
```

```python
# select price column
y = us_housing['Price']
```

## Split de treino

Agora vamos dividir os dados em um conjunto de treinamento e um conjunto de testes. Vamos criar o modelo usando o conjunto de treinamento e depois usar o conjunto de testes para avaliar o modelo.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

## Criando e treinando o modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
# instantiate the linear regression class
lm = LinearRegression()
```

```python
# fit/ train the model
lm.fit(X_train,y_train)
```

```python
r2_train = lm.score(X_train, y_train)
r2_test = lm.score(X_test, y_test)
print('R2 no set de treino: %.2f' % r2_train)
print('R2 no set de teste: %.2f' % r2_test)
```

* The definition of R-squared is fairly straight-forward; it is the percentage of the response variable variation that is explained by a linear model.

# exemplo stasmodel

Vamos começar com as vantagens dos statsmodels sobre o scikit-learn. O Scikit-Learn não é feito para estatísticas hardcore. Ele lhe dará todas as funcionalidades básicas, mas não lhe dará as mais sofisticadas que os estatísticos profissionais geralmente usam. Um exemplo simples seria o sci-kit learn fornecer uma infinidade de diferentes métodos de ajuste linear, mas apenas os parâmetros de ajuste, R ^ 2, MSE, etc. Mas se você precisar de outros valores de diagnóstico, como p-values, erros padrão em cada um dos parâmetros, t-tests, etc. Então você tem que escrever seu próprio código ou mudar para statsmodels (que foi a razão para eu mover). Além disso, a sintaxe dos statsmodels está mais alinhada com a do R, então se você for um usuário de R e quiser mudar para python, os statsmodels serão mais fáceis de aprender.

A maior desvantagem dos statsmodels reside no fato de que ainda é um pacote muito novo. A quantidade e qualidade da documentação disponível é muito pobre. Por outro lado, o scikit-learn tem uma documentação muito bem escrita. O Scikit-learn usa uma abordagem modular simples para todas as suas funções (fit-> transform-> predict) e, portanto, é muito fácil de aprender e se torna muito intuitivo depois que você conhece o básico. É mais "pythônico" do que os modelos de estatísticas. Além disso, há muitos bugs e objetos parcialmente implementados em statsmodels (não é que não haja nenhum em scikit-learn, mas eles são menos numerosos), então às vezes você tem que usá-lo com cuidado adequado. A maioria desses problemas, como documentação / bugs, desaparecerá conforme novas versões forem lançadas.

Então, para concluir, se você pretende usar os pacotes para tópicos relacionados a aprendizado de máquina, o scikit-learn tem muito mais recursos disponíveis do que statsmodels, mas se você pretende usá-los para análise estatística, então você deve ir para statsmodels.

## Métricas de avaliação de regressão


Aqui estão três métricas de avaliação comuns para problemas de regressão:

**Mean absolute error ** (erro absoluto médio) (MAE) é a média do valor absoluto dos erros:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

** Mean Squared Error ** (erro médio quadrático) (MSE) é a média dos erros quadrados:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

** Root Mean Square Error ** (raiz do erro quadrático médio) (RMSE) é a raiz quadrada da média dos erros quadrados:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparando estas métricas:

- **MAE** é o mais fácil de entender, porque é o erro médio.
- **MSE** é mais popular que o MAE, porque a MSE "puniria" erros maiores, o que tende a ser útil no mundo real.
- **RMSE** é ainda mais popular do que MSE, porque o RMSE é interpretável nas unidades "y".

Todas estas são ** funções de perda **, porque queremos minimizá-las.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
